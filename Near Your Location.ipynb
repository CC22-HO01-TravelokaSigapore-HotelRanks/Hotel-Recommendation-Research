```python
import pandas as pd
import numpy as np
import haversine as hs
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df_hotel = pd.read_csv('/content/drive/Shareddrives/CapBatu/ML Divsion/Data Scraping/data/hotel_dummy_photos.csv')
df_hotel = df_hotel[['name','latitude','longitude','hotel_star']]
```

```python
# Concatenating latitude and longitude to create a consolidated location as accepted by havesine function
df_hotel['coor'] = list(zip(df_hotel.latitude, df_hotel.longitude))
```

```python
df_hotel.head()
```

```python
# Function to calculate distance between two locations (hotel and user)
# loc1 = location of the user
# loc2 = location of a hotel
import geocoder
g = geocoder.ip('me')

def distance_from(loc1,loc2): 
    distance = hs.haversine(loc1,loc2)
    return distance

user_coor = (-8.685924978406074, 115.16562524113574)

distances_km = []
for row in df_hotel.itertuples(index=False):
   distances_km.append(distance_from(user_coor, row.coor))

df_hotel['distance_from_user'] = distances_km
```

```python
df_hotel.sort_values(['distance_from_user', 'hotel_star'], ascending=[True, False])
```
