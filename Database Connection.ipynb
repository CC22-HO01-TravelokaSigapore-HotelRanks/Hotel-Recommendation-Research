```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
db_connect_string=''
engine = create_engine(db_connect_string, connect_args={'connect_timeout': 5})
df = pd.read_sql("SELECT * FROM hotel_dummy ", engine)
df
```

```python
def db_evaluate(df_in: pd.DataFrame) -> pd.DataFrame:
  df = df_in.copy()
  cols = df.columns.to_list()
  dtypes = df.dtypes
  for i in cols:
    if dtypes[i] == "O":
      try:
        df[i] = df[i].map(pd.eval)
      except:
        print(f"Columns '{i}' cannot be evaluated")
```
